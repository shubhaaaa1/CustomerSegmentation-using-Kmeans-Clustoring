Splitting the Important Features 


```python
X = data.iloc[:,[3,4]].values
```

```python
print(X)
```

Choosing the Number of clusters Using WCSS -> within cluster sum of square


```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)

    wcss.append(kmeans.inertia_)
    
```

Elbow Point Graph of the Clusters

```python
sns.set_theme()
plt.figure(figsize=(4,4))
plt.plot(range(1,11),wcss)
plt.title('Elbow Point Graph : ')
plt.xlabel('No. of Clustors : ')
plt.ylabel('WCSS ')
plt.show()
```

Optimal number of Clustors is 5 because out of two elbow points, this one is the last

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=0)
Y= kmeans.fit_predict(X)
print(Y)
```

Cluster Visualization 

```python
plt.figure(figsize=(5,5))
plt.scatter(X[Y==0,0],X[Y==0,1],s=40,c='green',label = 'Clustor 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=40,c='red',label = 'Clustor 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=40,c='yellow',label = 'Clustor 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=40,c='black',label = 'Clustor 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=40,c='blue',label = 'Clustor 5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='Centroids')

plt.title('Customer Segmentation : ')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score ')
plt.show()
```
